```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
mean1,sigma1=5,7
mean2,sigma2=80,10
d1=np.random.normal(mean1,sigma1,1000)
d2=np.random.normal(mean2,sigma2,1000)
sns.kdeplot(d1,
            linewidth=3,
            color="b",
            label="distribution 1");
sns.kdeplot(d2,
             linewidth=3,
             color="r",
             label="distribution 2");
plt.legend()
```

```python
sns.kdeplot((d1-d1.mean())/d1.std(),color="b",label="standard distribution 1");
sns.kdeplot((d2-d2.mean())/d2.std(),color="r",label="standard distribution 2");
plt.legend()
```
